## Завдання 1. Агент з інструментами веб-пошуку

Створюємо агента рівня L1 з підключеним інструментом DuckDuckGoTools для пошуку в інтернеті.
Агент відповідає українською мовою та вказує джерела інформації.

```python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.tools.duckduckgo import DuckDuckGoTools

# Створення агента з веб-пошуком
web_agent = Agent(
    name="Web Research Agent",
    model=Gemini(id="gemini-2.5-flash"),
    tools=[DuckDuckGoTools()],
    instructions=[
        "Відповідай українською мовою.",
        "Завжди вказуй джерела інформації.",
        "Надавай структуровані та інформативні відповіді.",
        "ОБОВ'ЯЗКОВО використовуй інструменти пошуку для кожного запиту.",
    ],
    debug_mode=True,
    markdown=True,
)

# Запит 1 — актуальні новини
print("=" * 50)
print("ЗАПИТ 1: Новини про ChatGPT та Claude AI")
print("=" * 50)
web_agent.print_response(
    "Знайди останні новини про ChatGPT та Claude AI.",
    stream=True,
    stream_intermediate_steps=True
)

# Запит 2 — факти про технології
print("=" * 50)
print("ЗАПИТ 2: Найновіші смартфони")
print("=" * 50)
web_agent.print_response(
    "Знайди інформацію про найновіші смартфони та їх характеристики.",
    stream=True,
    stream_intermediate_steps=True
)

# Запит 3 — новини про Україну
print("=" * 50)
print("ЗАПИТ 3: Технологічні новини України")
print("=" * 50)
web_agent.print_response(
    "Знайди новини про розвиток IT індустрії в Україні.",
    stream=True,
    stream_intermediate_steps=True
)
```

## Завдання 2. Власні інструменти

Створюємо два власні інструменти за допомогою декоратора `@tool`:
1. `get_current_datetime` — повертає поточну дату та час
2. `check_number` — перевіряє число на парність та чи є воно простим

```python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.tools import tool

@tool
def get_current_datetime(timezone: str = "Europe/Kyiv") -> str:
    """Отримує поточну дату та час у вказаному часовому поясі.

    Args:
        timezone: Часовий пояс (за замовчуванням Europe/Kyiv).

    Returns:
        Рядок з поточною датою та часом у форматі ISO.
    """
    from datetime import datetime
    import pytz
    tz = pytz.timezone(timezone)
    now = datetime.now(tz)
    return f"Поточна дата та час ({timezone}): {now.strftime('%Y-%m-%d %H:%M:%S')}"

@tool
def check_number(number: int) -> str:
    """Перевіряє число на парність та чи є воно простим числом.

    Args:
        number: Ціле число для перевірки.

    Returns:
        Рядок з результатами перевірки числа.
    """
    is_even = number % 2 == 0
    is_prime = number > 1 and all(number % i != 0 for i in range(2, int(number**0.5) + 1))
    return (
        f"Число {number}:\n"
        f"- {'Парне' if is_even else 'Непарне'}\n"
        f"- {'Просте' if is_prime else 'Не просте'}"
    )

# Створення агента з власними інструментами
custom_agent = Agent(
    name="Custom Tools Agent",
    model=Gemini(id="gemini-2.5-flash"),
    tools=[get_current_datetime, check_number],
    instructions=[
        "Використовуй інструменти для відповіді.",
        "Відповідай українською мовою.",
    ],
    debug_mode=True,
    markdown=True,
)

# Тест 1 — дата та час
print("=" * 50)
print("ТЕСТ 1: Поточний час")
print("=" * 50)
custom_agent.print_response(
    "Котра зараз година в Києві?",
    stream=True,
    stream_intermediate_steps=True
)

# Тест 2 — перевірка числа
print("=" * 50)
print("ТЕСТ 2: Перевірка числа")
print("=" * 50)
custom_agent.print_response(
    "Перевір число 17 — чи воно парне і чи є простим?",
    stream=True,
    stream_intermediate_steps=True
)
```

### Висновок до Завдання 2
Створено два власні інструменти з детальними docstrings. Агент коректно викликає
`get_current_datetime` для отримання часу та `check_number` для перевірки числових
властивостей. Docstring є критично важливим — без нього модель не знає коли викликати інструмент.

## Завдання 3. Команда агентів (Team)

Створюємо мультиагентну систему з двох спеціалізованих агентів:
- **Researcher** — збирає інформацію з інтернету
- **Analyst** — аналізує та структурує отримані дані

Агенти працюють у режимі `coordinate` — послідовна координація.

```python
from agno.agent import Agent
from agno.team.team import Team
from agno.models.google import Gemini
from agno.tools.duckduckgo import DuckDuckGoTools

# Агент-дослідник
researcher = Agent(
    name="Researcher",
    role="Збір та верифікація інформації з інтернету",
    model=Gemini(id="gemini-2.5-flash"),
    tools=[DuckDuckGoTools()],
    instructions=[
        "Шукай актуальну інформацію з надійних джерел.",
        "Завжди вказуй джерела.",
        "Відповідай українською мовою."
    ],
    markdown=True
)

# Агент-аналітик
analyst = Agent(
    name="Analyst",
    role="Аналіз інформації та написання структурованих звітів",
    model=Gemini(id="gemini-2.5-flash"),
    instructions=[
        "Аналізуй отриману інформацію.",
        "Структуруй відповідь з чітким вступом та висновками.",
        "Пиши лаконічно та інформативно українською."
    ],
    markdown=True
)

# Створення команди
research_team = Team(
    name="Research Team",
    members=[researcher, analyst],
    model=Gemini(id="gemini-2.5-flash"),
    instructions=[
        "1. Спочатку попроси Researcher знайти інформацію.",
        "2. Потім передай результати Analyst для аналізу.",
        "3. Поверни структурований звіт українською мовою."
    ],
    share_member_interactions=True,
    show_members_responses=True,
    markdown=True,
    debug_mode=True
)

# Тестування команди
research_team.print_response(
    "Досліди останні тренди у сфері generative AI та напиши короткий аналітичний звіт.",
    stream=True
)
```

### Висновок до Завдання 3
Команда агентів успішно створена в режимі `coordinate`. Координатор делегує задачі
через `delegate_task_to_member` — спочатку Researcher збирає інформацію, потім
Analyst аналізує результати. `share_member_interactions=True` забезпечує передачу
контексту між агентами.

## Завдання 4. Агент з можливостями Reasoning

Створюємо агента з `ReasoningTools` для покрокового міркування.
Агент аналізує складні задачі крок за кроком перед тим як дати відповідь.

```python
from agno.agent import Agent
from agno.models.google import Gemini
from agno.tools.reasoning import ReasoningTools

# Створення агента з reasoning
reasoning_agent = Agent(
    name="Reasoning Agent",
    model=Gemini(id="gemini-2.5-flash"),
    tools=[
        ReasoningTools(
            enable_think=True,
            enable_analyze=True,
            add_instructions=True
        )
    ],
    instructions=[
        "Перед відповіддю аналізуй проблему покроково.",
        "Розглядай різні перспективи та можливі наслідки.",
        "Чітко формулюй висновки з обґрунтуванням.",
        "Відповідай українською мовою."
    ],
    markdown=True,
    debug_mode=True
)

# Складне запитання для reasoning
reasoning_agent.print_response(
    """
    Уяви, що ти консультант стартапу в сфері EdTech.
    Команда вирішує між двома стратегіями розвитку:
    А) Швидко випустити базовий MVP для 1000 студентів
    Б) Витратити 6 місяців на створення досконалої платформи з AI-функціями

    Які фактори слід врахувати? Яку стратегію ти б рекомендував і чому?
    """,
    stream=True,
    stream_intermediate_steps=True
)
```

### Висновок до Завдання 4
Агент з `ReasoningTools` демонструє покрокове міркування через інструменти `think` та `analyze`.
Видно три кроки мислення: аналіз обох стратегій, виділення факторів для кожної.
Такий підхід забезпечує обґрунтовані відповіді на складні аналітичні запитання.
